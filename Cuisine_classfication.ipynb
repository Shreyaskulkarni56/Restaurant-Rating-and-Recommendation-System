# Import libraries

```python
# import libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#load the data set
df = pd.read_csv("Dataset .csv")
df
```

  # Data Preprocessing

```python
#remove features
columns_to_drop = [
    'Restaurant ID', 'Country Code', 'City', 'Address', 'Locality',
    'Locality Verbose', 'Longitude', 'Latitude', 'Currency',
    'Has Table booking', 'Has Online delivery', 'Is delivering now',
    'Switch to order menu', 'Price range', 'Aggregate rating',
    'Rating color', 'Rating text', 'Votes'
]

df.drop(columns=columns_to_drop, axis=1, inplace=True)
```

```python
df.isnull().sum()
```

```python
df.dropna(inplace=True)
```

```python
df.shape
```

```python
df.describe(include="all")
```

```python
from sklearn.preprocessing import LabelEncoder

label_encoder = LabelEncoder()

# Encode selected columns
columns_to_encode = ['Restaurant Name', 'Cuisines']
for col in columns_to_encode:
    df[col] = label_encoder.fit_transform(df[col])

df
```

```python
X= df.drop('Cuisines',axis=1)
Y= df['Cuisines']
```

```python
from sklearn.preprocessing import StandardScaler 
```

```python
scaler = StandardScaler()
X=scaler.fit_transform(X)
```

```python
from sklearn.model_selection import train_test_split

X_train,X_test,Y_train,Y_test =train_test_split(X,Y,test_size=0.2,random_state=15)
```

# Random Forest Model

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, accuracy_score


# Step 1: Initialize the model
model_rfc = RandomForestClassifier(n_estimators=100, random_state=42)

# Step 2: Train the model
model_rfc.fit(X_train, Y_train)

# Step 3: Make predictions
rfc_pred = model_rfc.predict(X_test)

# Step 4: Evaluate the model
print(" Accuracy:", accuracy_score(Y_test, rfc_pred))
print("\n Classification Report:\n", classification_report(Y_test, rfc_pred))
```

# LogisticRegression

```python
from sklearn.linear_model import LogisticRegression
```

```python
logreg=LogisticRegression(multi_class="multinomial")
logreg.fit(X_train,Y_train)
```

```python
logreg_prediction=logreg.predict(X_test)
print(logreg_prediction)
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

def evaluate_model(Y_test, Y_train, model_name="Model"):
    accuracy = accuracy_score(Y_test, Y_train)
    precision = precision_score(Y_test, Y_train, average='micro')
    recall = recall_score(Y_test, Y_train, average='micro')
    f1 = f1_score(Y_test, Y_train, average='micro')

    print(f" Evaluation Results for {model_name}:")
    print(f" Accuracy:  {accuracy:.2f}")
    print(f" Precision: {precision:.2f}")
    print(f" Recall:    {recall:.2f}")
    print(f" F1-score:  {f1:.2f}")

# Example usage:
evaluate_model(Y_test, logreg_prediction, model_name="Logistic Regression")
```
